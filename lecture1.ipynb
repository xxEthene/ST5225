```python
import pandas as pd

data = {
    "node": ["A", "B", "C", "D"],
    "degree": [3, 2, 5, 1]
}
df = pd.DataFrame(data)
display(df)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["node"], df["degree"])
plt.xlabel("Node")
plt.ylabel("Degree")
plt.title("Node Degrees")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [("A", "B"), ("A", "C"), ("C", "D"), ("A", "D")]
G.add_edges_from(edges)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1200, font_size=14)
plt.title("Simple Graph")
plt.show()
```

```python
# Get degrees for each node
degrees = dict(G.degree())
# Normalize node sizes (e.g., multiply degree by a factor for visibility)
node_sizes = [degrees[node] * 800 for node in G.nodes()]
# Color nodes by degree (higher degree = darker blue)
node_colors = [degrees[node] for node in G.nodes()]

pos = nx.spring_layout(G)
nx.draw(
    G, pos, with_labels=True,
    node_color=node_colors,
    cmap=plt.cm.Blues,
    edge_color='gray',
    node_size=node_sizes,
    font_size=14
)
plt.title("Simple Graph (Node color/size by degree)")
plt.show()
```

```python
G = nx.Graph()
edges = [("A", "B"), ("A", "C"), ("C", "D"), ("A", "D")]
G.add_edges_from(edges)

print("Nodes:", list(G.nodes()))
print("Edges:", list(G.edges()))
```

```python
# Get degrees from the graph
degree_dict = dict(G.degree())
df = pd.DataFrame({
    "node": list(degree_dict.keys()),
    "degree": list(degree_dict.values())
})
df = df.sort_values(by="degree", ascending=False)
display(df)
```

```python
plt.bar(df["node"], df["degree"])
plt.xlabel("Node")
plt.ylabel("Degree")
plt.title("Node Degrees")
plt.show()
```
